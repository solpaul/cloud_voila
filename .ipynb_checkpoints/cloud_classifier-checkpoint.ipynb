# What's that cloud?

Look up! Can you tell the difference between a cumulus and cumulonimbus, stratus and altostratus, or cirrus and cirrostratus? If not then you've come to the right place. Take a photo of the sky, and click 'upload' to classify it.

----

```python
cloud_name_dict = {
    'Ac': 'Altocumulus',
    'Sc': 'Stratocumulus',
    'Ns': 'Nimbostratus',
    'Cu': 'Cumulus',
    'Ci': 'Cirrus',
    'Cc': 'Cirrocumulus',
    'Cb': 'Cumulonimbus',
    'As': 'Altostratus',
    'Ct': 'Contrail',
    'Cs': 'Cirrostratus',
    'St': 'Stratus',
}
```

```python
def parent_label_mapped(o):
    return cloud_name_dict[Path(o).parent.name]
```

```python
path = Path()
learn_inf = load_learner(path/'export.pkl', cpu=True)
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
def on_click(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img)
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'This looks like a {pred} (probability {probs[pred_idx]:.04f})'
```

```python
btn_upload.observe(on_click, names=['data'])
```

```python
display(VBox([widgets.Label('Select your cloud!'), btn_upload, widgets.HTML('<br>'), out_pl, lbl_pred]))
```
